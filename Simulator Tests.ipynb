```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd
```

```python
TVO_v1 = pd.read_csv('Simulator/TVO_v1.csv')
print(TVO_v1)
```

```python
finished_counts_tvo_v1 = TVO_v1['finished'].value_counts()
print(finished_counts_tvo_v1/2)
vitamins_used = TVO_v1['vitamins_used'].mean()
print(vitamins_used)
```

```python
finished_counts_tvo_v1.plot(kind='pie')
```

```python
TVO_v2 = pd.read_csv('Simulator/TVO_v2.csv')
finished_counts_tvo_v2 = TVO_v2['finished'].value_counts()
print(finished_counts_tvo_v2)
print(TVO_v2['vitamins_used'].mean())
```

```python
finished_counts_tvo_v2.plot(kind='pie')
```

```python
TVO_v3 = pd.read_csv('Simulator/TVO_v3.csv')
finished_counts_tvo_v3 = TVO_v3['finished'].value_counts()
print(finished_counts_tvo_v3)
print(TVO_v3['vitamins_used'].mean())
finished_counts_tvo_v3.plot(kind='pie')
```

```python
TVO_v4 = pd.read_csv('Simulator/TVO_v4.csv')
finished_counts_tvo_v4 = TVO_v4['finished'].value_counts()
print(finished_counts_tvo_v4)
print(TVO_v4['vitamins_used'].mean())

print(finished_counts_tvo_v2[1])
```

```python
threshold = int(len(TVO_v1)/2)

df = pd.DataFrame({'ver':['V1', 'V2', 'V3', 'V4'], 'val':[int(finished_counts_tvo_v1[1]/2), int(finished_counts_tvo_v2[1]/2), int(finished_counts_tvo_v3[1]/2),int(finished_counts_tvo_v4[1]/2)]})
ax = df.plot.bar(x='ver', y='val', rot=0, color=['mediumslateblue', 'mediumslateblue', 'mediumslateblue', 'mediumslateblue'])
ax.get_legend().remove()
ax.set_ylabel('# of Levels')
ax.set_xlabel('Version')
ax.set_facecolor('black')
ax.set_title('Number of Levels Finished by the TVO Solvers')

for i in range(len(df['ver'])):
    plt.text(i, df['val'][i]//1.2, df['val'][i], ha = 'center',
                 Bbox = dict(facecolor = 'white'))
    
plt.axhline(y=threshold,linewidth=6, color='y', linestyle = '--', label = ("All Puzzles = "+str(threshold)))

    
plt.figure(figsize=(30,15)) 
plt.rcParams["figure.figsize"] = (30,15)
plt.rcParams.update({'font.size': 50})
#     plt.text(i, df['val'][i], df['val'][i], ha = 'center',color='white')
```

```python
TVO_v1_success = TVO_v1[TVO_v1['finished']==1]
TVO_v2_success = TVO_v2[TVO_v2['finished']==1]
TVO_v3_success = TVO_v3[TVO_v3['finished']==1]
TVO_v4_success = TVO_v4[TVO_v4['finished']==1]

df = pd.DataFrame({'ver':['V1', 'V2', 'V3', 'V4'], 'val':[round(TVO_v1_success['vitamins_used'].mean(),2), round(TVO_v2_success['vitamins_used'].mean(),2),round(TVO_v3_success['vitamins_used'].mean(),2),round(TVO_v4_success['vitamins_used'].mean(),2)]})
ax = df.plot.bar(x='ver', y='val', rot=0,color=['mediumslateblue', 'mediumslateblue', 'mediumslateblue', 'mediumslateblue'])

ax.get_legend().remove()
ax.set_ylabel('Mean # of Vitamins')
ax.set_xlabel('Version')
ax.set_facecolor('black')
ax.set_title('Mean # of Vitamins used in Puzzles Finished by the TVO Solvers')


for i in range(len(df['ver'])):
    plt.text(i, df['val'][i]//1.2, df['val'][i], ha = 'center',
                 Bbox = dict(facecolor = 'white'))

plt.figure(figsize=(25,15)) 
plt.rcParams["figure.figsize"] = (20,15)
```
